```python
a = float(input('enter the value of a:- '))
b = float(input('enter the value of b:- '))
c = float(input('enter the value of c:- '))
x0 = float(input('enter the value of x0:- '))
v0 = float(input('enter the value of v0:- '))
```

```python
import math
import plotly.express as px
```

```python
alpha = -b/(2*a)
bita = pow(abs(b**2 - 4*a*c),0.5)/(2*a)
l1 = alpha + bita
l2 = alpha - bita
if l1 != l2:
  d1 = (v0 - x0*l2)/(l1 - l2)
  d2 = (v0 - x0*l1)/(l2 - l1)
# print(bita,l1,l2,alpha)
```

```python
t = [i/100 for i in range(1001)]

x_lst = []
```

```python
d = b**2 - 4*a*c
for i in t:
  if d > 0:
    # print('over damped')
    x = d1*math.exp(l1*i) + d2*math.exp(l2*i)
    x_lst.append(x)
  elif d == 0:
    # print('critical damped')
    x = (x0 + (v0 - (alpha)*x0)*i)*math.exp((-b/2*a)*i)
    x_lst.append(x)
  else:
    # print('under damped')
    x = math.exp((-alpha)*i)*(x0*math.cos(bita*i) + ((v0 + alpha*x0)/bita)*math.sin(bita*i))
    x_lst.append(x)
```

```python
if d > 0:
  print('it is over damping')
elif d == 0:
  print('it is critical damping')
else:
  print('it is under damping')
```

```python
fig = px.line(x=t, y = x_lst)
fig.show()
```
